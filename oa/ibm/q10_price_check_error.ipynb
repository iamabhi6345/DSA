```python
def f(p,pp,ps,sp):
    d = dict()
    for i in range(len(p)):
        d[p[i]] = pp[i]
    ans =0
    for i in range(len(ps)):
        if d[ps[i]]!=sp[i]:
            ans+=1
    return ans
```

```python
p = ['eggs', 'milk', 'cheese'] 
pp = [2.89, 3.29, 5.79] 
ps=['eggs', 'eggs', 'cheese', 'milk'] 
sp=[2.89, 2.99, 5.97, 3.29]

print(f(p,pp,ps,sp))
```
