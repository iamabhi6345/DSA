```python
def counter(arr):
    n = len(arr)
    ctr=[0]*n 
    for i in range(n):
        diff=0
        for j in range(i):
            if arr[j]<arr[i]:
                diff += abs(arr[j]-arr[i])
            elif arr[j]>arr[i]:
                 diff -= abs(arr[i]-arr[j])
        ctr[i]=diff
    
    return ctr
                
```

```python
print(counter([2,4,3]))
print(counter([2,1,3]))
```
