```python
def f(ind , arr , tmp , d):
    if tmp !=None and len(tmp)==3:
        s=sum(tmp)
        if s%d==0:
            print(tmp)
            return 1
        
        return 0
    if ind>=len(arr):
        return 0
    
    
    
    take = f(ind+1 , arr , tmp+[arr[ind]] , d)
 
    nottake= f(ind+1 , arr , tmp , d)
    return take+nottake
```
