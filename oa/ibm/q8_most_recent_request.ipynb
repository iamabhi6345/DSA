```python
def f(ls , k):
    st = set()
    ans=list()
    n = len(ls)
    for i in range(n-1 , -1 , -1):
        if len(st)==k:
            return ans  
        if ls[i] not in st:
            ans.append(ls[i])
            st.add(ls[i])
    if len(st)==k:
        return ans   
    return -1
```

```python
ls=['i1' , 'i2' , 'i3' , 'i1' , 'i3']
print(f(ls , 3))
```
