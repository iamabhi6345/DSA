```python
""" 
to make minimum length every element frequency should be 1
freq=1 --> no need to delete anything
freq=2 --> delete 1 element so 1 move neded
freq=3 --> delete middle one one left so 1 move required
freq=4 (1,2,3,4)--> choose 3 -->2,4 will be deleted ,left=1,3 take one more move 
        so now 1 left total 2 move
same for others logic

observation move =freq//2
"""
```

```python
def minimummove(s):
    tmp = dict()
    for i in s:
        if i not in tmp:
            tmp[i]=1
        else:
            tmp[i]+=1
    
    count=0
    for i in tmp:
        count+=tmp[i]//2
    
    return count
```

```python
print(minimummove("baabacaa"))
print(minimummove("baabacaa"))
```
