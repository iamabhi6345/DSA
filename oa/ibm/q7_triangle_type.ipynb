```python
def valid(a,b,c):
    l=sorted([a,b,c])
    if l[2]<(l[0]+l[1]):
        return True 
    return False 

def f(ls):
    ss=[]
    for s in ls:
        a,b,c=map(int , s.split(" "))
        if valid(a,b,c):
            if a==b==c:
                ss.append("Equilateral")
            elif (a==b) or (a==c) or (b==c):
                ss.append("isosceles")
            else:
                ss.append("None of these")
        else:
                ss.append("None of these")
    
    return ss
        
```

```python
ls=['2 2 1','3 3 3','3 4 5','1 1 3']
print(f(ls))
```
