```python
from collections import defaultdict

def find_missing_prisoner(locations):
    x_count = defaultdict(int)
    y_count = defaultdict(int)
    
    for x, y in locations:
        x_count[x] += 1
        y_count[y] += 1
    
    missing_x = None
    missing_y = None
    
    for x in x_count:
        if x_count[x] % 2 != 0:
            missing_x = x
            break
    
    for y in y_count:
        if y_count[y] % 2 != 0:
            missing_y = y
            break
    
    return (missing_x, missing_y)


locations = [(1, 1), (1, 2), (2, 1), (4, 4), (4, 6), (9, 4), (9, 6)]
missing_prisoner = find_missing_prisoner(locations)
print(f"The missing prisoner's coordinates are: {missing_prisoner}")
```
