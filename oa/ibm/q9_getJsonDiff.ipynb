```python
def f( j1 , j2 ):
    ans =[]
    for i in j1:
        if i in j2:
            if j1[i]!=j2[i]:
                ans.append(i)
    ans.sort()
    return ans
```

```python
json1 = '{"hello": "world", "hi": "hello", "you": "me"}'
json2 = '{"hello": "world", "hi": "helloo", "you": "me"}'

import json
json1=json.loads(json1)
json2=json.loads(json2)
print(f(json1,json2))
```
